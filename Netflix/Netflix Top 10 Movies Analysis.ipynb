## Data Wrangling

#### Datanın sütunları ilə tanış olun. Onlar hansı data tiplərindədir? Bütün sütunların data tipləri uyğundur?

```python
df.info()
```

#### Düzgün data tipində olmayan müvafiq data tiplərinə keçirin

```python
df['As of'] = pd.to_datetime(df['As of'], errors='coerce')
df['Netflix Release Date'] = pd.to_datetime(df['Netflix Release Date'], errors='coerce')
df['Year to Date Rank'] = pd.to_numeric(df['Year to Date Rank'], errors='coerce')
df['Last Week Rank'] = pd.to_numeric(df['Last Week Rank'], errors='coerce')
```

```python
df.info()
```

```python
df.head()
```

#### Hansı sütunlarda boş sətrlər var? Hansı sütunlarda neçə boş sətr var?

```python
df.isnull().sum()
```

#### Boş sətrlər olan sütunda olan bütün boş məlumatları "No" yazısı ilə doldurun

```python
df['Year to Date Rank']=df['Year to Date Rank'].fillna(value=0).astype('int')
df['Last Week Rank']=df['Last Week Rank'].fillna(value=0).astype('int')
df['Netflix Exclusive']=df['Netflix Exclusive'].fillna(value='No')
```

```python
df.info()
```

```python
df.head()
```

#### Yeni bir dashboard yaradın (15,15 ölçüsündə). Ən önəmli gördüyünüz 4 ədəd vizualı bu dashboard-da göstərin

```python
df.head(2)
```

```python
plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
df1 = df.groupby('Title')['Days In Top 10'].max().reset_index().sort_values(by='Days In Top 10',ascending=False).head(10)
sns.barplot(y='Title', x='Days In Top 10', data=df1, palette='viridis')
plt.title('En cox ilk 10-luga giren filmler')

plt.subplot(2,2,2)
sns.countplot(x=df['Netflix Release Date'].dt.year, data=df, palette='crest')
plt.title('Netflixde buraxilma illerine gore filmlerin sayi')

plt.subplot(2,2,3)
df3 = df.groupby(df['Netflix Release Date'].dt.year)['Viewership Score'].mean().reset_index()
sns.barplot(y='Viewership Score', x='Netflix Release Date', data=df3, palette='coolwarm')
plt.title('Netflixde buraxilma illerine gore ortalama istifadeci reytingi')

plt.subplot(2,2,4)
df4 = df.groupby(['Aylar','Type'])['Viewership Score'].mean().reset_index()
for t in df4['Type'].unique():
    subset = df4[df4['Type'] == t]
    plt.plot(subset['Aylar'], subset['Viewership Score'], marker='o', label=t)
plt.title("Aylar uzre film tiplerine gore ortalama istifadeci reytingi")
plt.xticks(rotation=45)
plt.legend()

plt.show()
```
